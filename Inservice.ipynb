## Coordinate System
### When we draw shapes, the libraries we use invert the y-axis from the typical coordinate plane. Otherwise, we use coordinate pairs (x, y) in the same manner as in an algebra class.

```python
# The matplotlib library is for creating graphs and other visuals
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0, 600)
ax.set_ylim(0, 600)
ax.invert_yaxis()

# Draw a square with a top left corner at (100, 200)
square = patches.Rectangle((100, 200), 100, 100)
ax.add_patch(square)

# Display the plot
plt.show()
```

## Trigonometry
### We convert polar coordinates to rectangular coordinates whenever we want to use angles to dictate the locations of points

```python
import matplotlib.pyplot as plt
import numpy as np

# Choose number of points and generate angles in radians
num_points = 12
angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)


# Convert polar coordinates to Cartesian coordinates
x = 300 + 200 * np.cos(angles)
y = 300 + 200 * np.sin(angles)

# Create a scatter plot in a rectangular coordinate system
fig, ax = plt.subplots(figsize=(6, 6))
plt.scatter(x, y, color='red', marker='o')  # Invert the y-axis
ax.set_xlim(0, 600)
ax.set_ylim(0, 600)
ax.invert_yaxis()

plt.show()
```
